```python
# 折线图

from turtle import color
from matplotlib import pyplot as plt
import random

x = range(121)
y = [random.randint(18,40) for i in range(121)]
y2 = [random.randint(18,40) for i in range(121)]

plt.figure(figsize=(15,4),dpi=80) #设置图形大小，清晰度

plt.plot(x,y,label='A',color='r') # lable图例名字
plt.plot(x,y2,label='B',color='b')


t = list(range(121)) # x步长
l = [f'this is {i}' for i in range(121)] # x步长对应的label
plt.xticks(t[::10],l[::10],rotation=40)  # 旋转
plt.yticks(range(15,45,5))

# plt.xlabel('Time')
# plt.ylabel('Temperature')
# plt.title('Prediction')
plt.grid(alpha=0.2) # 画网格

plt.legend(loc='best') # 图例
plt.show()
```
